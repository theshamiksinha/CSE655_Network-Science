```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def in_out_deg(filepath, num_nodes):
    indegree = []
    outdegree = []
    for i in range(num_nodes):
        indegree.append(0)
        outdegree.append(0)
    
    with open(filepath, 'r') as file:
        for line in file:
            nodes = line.strip().split()
            src, dest = map(int, nodes)
            outdegree[src] += 1
            indegree[dest] += 1
    
    return indegree, outdegree


def plot_degree_distribution(degrees, title):
    degree_counts = Counter(degrees)
    k_vals = np.array(list(degree_counts.keys()))
    p_k = np.array(list(degree_counts.values())) / sum(degree_counts.values())
    plt.figure(figsize=(8, 6))
    plt.scatter(k_vals, p_k, marker='o', color='b', label="Degree Distribution")
    #plt.xscale("log")
    #plt.yscale("log")
    plt.xlabel("Degree k")
    plt.ylabel("p(k)")
    plt.title(title)
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

# plotting the graphs showcasing the probabilities of a node having an indegree and outdegree equal to k
indegree, outdegree = in_out_deg('email-Eu-core.txt', 1005)
plot_degree_distribution(indegree, "In-degree Distribution")
plot_degree_distribution(outdegree, "Out-degree Distribution")
```

```python
# examples showing indegree and outdegree for some nodes
print(indegree[0])
print(outdegree[0])
print(indegree[27])
print(outdegree[27])
```

```python
# plotting the frequency of nodes having a specific indegree and outdegree
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.hist(indegree, bins=range(min(indegree), max(indegree) + 2))
plt.xlabel("Indegree")
plt.ylabel("Frequency")
plt.title("Indegree Distribution")
plt.subplot(1, 2, 2)
plt.hist(outdegree, bins=range(min(outdegree), max(outdegree) + 2))
plt.xlabel("Outdegree")
plt.ylabel("Frequency")
plt.title("Outdegree Distribution")
plt.tight_layout()
plt.show()
```
