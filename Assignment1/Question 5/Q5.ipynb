```python
from collections import defaultdict

adj_list = defaultdict(list)

with open("karate_club_edgelist.txt", "r") as f:
    for line in f:
        line = line.strip()
        if line:
            u, v = map(int, line.split())

            # making an undirected graph
            adj_list[u].append(v)
            adj_list[v].append(u)

test_graph = dict(adj_list)

print(test_graph)
```

```python
from collections import deque

def bfs(adj, source):
    visited = set()
    distance = {}
    parent = {}
    layers = defaultdict(list)

    queue = deque()

    # initializing the source node
    queue.append(source)
    visited.add(source)
    distance[source] = 0 # Distance of the source node from itself is 0
    parent[source] = None
    layers[0].append(source)

    # BFS traversal
    while queue:
        u = queue.popleft()
        
        # Visiting every neighbouring node
        for v in adj[u]:
            if v not in visited:
                visited.add(v)
                queue.append(v)

                # updating distance and parent
                distance[v] = distance[u] + 1
                parent[v] = u
                layers[distance[v]].append(v)

    return distance, parent, dict(layers)


source_node = 0

distance, parent, layers = bfs(test_graph, source_node)

print("Distances from source:")
for node in distance:
    print(f"Node {node}: {distance[node]}")

print("\nParents in BFS Tree:")
for node in parent:
    print(f"Node {node}: {parent[node]}")

print("\nBFS Layers:")
for layer in sorted(layers):
    print(f"Layer {layer}: {layers[layer]}")
```

---

## Complexity Analysis

### Time Complexity

Let:

- **V** = number of vertices (nodes)  
- **E** = number of edges  

In Breadth-First Search (BFS):

- Each vertex is visited exactly once.
- Every edge is examined once during traversal.

Therefore, the time complexity of BFS is:

\[
O(V + E)
\]

---

### Space Complexity

The auxiliary space used in BFS includes:

- Queue used for traversal → \(O(V)\)
- Distance array → \(O(V)\)
- Parent array → \(O(V)\)
- Visited set → \(O(V)\)

Hence, the overall space complexity is:

\[
O(V)
\]
